# Feature Engineering

```python
dateUnique = data['dteday'].unique()

# create new columns for each day's mean temp, high and low.
for i in dateUnique:
  data.loc[data.dteday == i, ['temp_mean', 'temp_high', 'temp_low']] = (
      data[data['dteday'] == i].temp_c.mean(),
      data[data['dteday'] == i].temp_c.max(),
      data[data['dteday'] == i].temp_c.min()
    )
  
# create new columns for each day's mean feels-like temp, high and low.
for i in dateUnique:
  data.loc[data.dteday == i, ['feels_like_mean', 'feels_like_high', 'feels_like_low']] = (
      data[data['dteday'] == i].feels_like_c.mean(),
      data[data['dteday'] == i].feels_like_c.max(),
      data[data['dteday'] == i].feels_like_c.min()
    )
  
# get the mean users each month. Create columns that show the previous month's mean users
monthUnique = data['month'].unique()
yearUnique = data['year'].unique()

for i in yearUnique:
  for j in monthUnique:

    data.loc[(data.year == i) & (data.month == j), ['casual_month_mean', 'registered_month_mean']] = (
        data[(data['year'] == i) & (data['month'] == j)].casual.mean(),
        data[(data['year'] == i) & (data['month'] == j)].registered.mean()
    )

    # dont get previous month if jan 2011
    if (i == 11 and j == 1):
      data.loc[(data.year == i) & (data.month == j), ['casual_prev_month_mean', 'registered_prev_month_mean']] = (
          data[(data['year'] == i) & (data['month'] == j)].casual.mean(),
          data[(data['year'] == i) & (data['month'] == j)].registered.mean()
      )
    # go back a year if jan 2012
    elif (i == 12 and j == 1): 
      data.loc[(data.year == i) & (data.month == j), ['casual_prev_month_mean', 'registered_prev_month_mean']] = (
          data[(data['year'] == i-1) & (data['month'] == 12)].casual.mean(),
          data[(data['year'] == i-1) & (data['month'] == 12)].registered.mean()
      )
    else:
      data.loc[(data.year == i) & (data.month == j), ['casual_prev_month_mean', 'registered_prev_month_mean']] = (
          data[(data['year'] == i) & (data['month'] == j-1)].casual.mean(),
          data[(data['year'] == i) & (data['month'] == j-1)].registered.mean()
      )



# data[data['month'] == 1].head(5)
# data[data['month'] == 2].head(5)
# data[data['month'] == 3].head(5)
# data[data['month'] == 4].head(5)
# data[(data['month'] == 1) & (data['year'] == 12)].head(5)
```

```python
# calculate the growth from the same month the previous year
for j in monthUnique:
    if j != 12:
      data.loc[(data.year == 12) & (data.month == j), ['ytd_casual_growth', 'ytd_registered_growth']] = (
          # (((new_number - old_number)/old_number) * 100)
          # old_number = data[(data['year'] == 11) & (data['month'] == j)].casual_month_mean.unique()
          # new_number = data[(data['year'] == 12) & (data['month'] == j)].casual_month_mean.unique()
          (((data[(data['year'] == 12) & (data['month'] == j)].casual_month_mean.unique()[0] 
          - data[(data['year'] == 11) & (data['month'] == j)].casual_month_mean.unique()[0])
          /data[(data['year'] == 11) & (data['month'] == j)].casual_month_mean.unique()[0]) * 100),
          (((data[(data['year'] == 12) & (data['month'] == j)].registered_month_mean.unique()[0]
          - data[(data['year'] == 11) & (data['month'] == j)].registered_month_mean.unique()[0])
          /data[(data['year'] == 11) & (data['month'] == j)].registered_month_mean.unique()[0]) * 100)
      )


# create new column populated with the previous months value
for j in monthUnique:

  # go back a year if jan 2012
  if j == 1:
    data.loc[(data.year == 11) & (data.month == j), ['ytd_prev_casual_growth', 'ytd_prev_registered_growth']] = (
        data[(data['year'] == 11) & (data['month'] == 12)].ytd_casual_growth.unique()[0],
        data[(data['year'] == 11) & (data['month'] == 12)].ytd_registered_growth.unique()[0]
    )
  else:
    data.loc[(data.year == 12) & (data.month == j), ['ytd_prev_casual_growth', 'ytd_prev_registered_growth']] = (
        data[(data['year'] == 12) & (data['month'] == j-1)].ytd_casual_growth.unique()[0],
        data[(data['year'] == 12) & (data['month'] == j-1)].ytd_registered_growth.unique()[0]
    )
```

```python

model = keras.Sequential()

model.add(keras.layers.Dense(feature_count, activation=keras.layers.LeakyReLU(alpha=0.3)))
# model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(feature_count, activation=keras.layers.LeakyReLU(alpha=0.3)))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(feature_count, activation=keras.layers.LeakyReLU(alpha=0.3)))
model.add(keras.layers.Dropout(0.2))
model.add(keras.layers.Dense(16, activation=keras.layers.LeakyReLU(alpha=0.3)))
model.add(keras.layers.Dense(2, activation='relu'))

opt = keras.optimizers.Adam(learning_rate=0.01)
model.compile(optimizer = opt, loss = 'mse', metrics = ['accuracy'])

# stop when val_accuracy is above threshhold
class MyThresholdCallback(tf.keras.callbacks.Callback):
    def __init__(self, threshold):
        super(MyThresholdCallback, self).__init__()
        self.threshold = threshold

    def on_epoch_end(self, epoch, logs=None): 
        val_acc = logs["val_accuracy"]
        loss = logs["loss"]
        if (val_acc >= self.threshold and epoch >= 20):
            self.model.stop_training = True
callback = MyThresholdCallback(threshold=0.89)

# es_callback = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)
model.fit(X_train, y_train, validation_data =(X_dev, y_dev), epochs = 60, callbacks=[callback])
```
